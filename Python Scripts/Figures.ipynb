```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec

# Confusion matrices (raw counts)
cm_1 = np.array([[220, 5], [5, 221]])   # GPT-4.1-mini ZS
cm_2 = np.array([[156, 37], [29, 165]]) # CLIP Few-Shot
cm_3 = np.array([[43, 3], [8, 37]])     # CNN
cm_4 = np.array([[224, 1], [4, 222]])   # GPT-4.1 Few-Shot CoT

cms = [cm_1, cm_2, cm_3, cm_4]
titles = [
    "Reasoning-based VLM\n(GPT-4.1-mini, ZS Prompt 1)",
    "Caption-based VLM\n(ViT-H-14 CLIP, Few-Shot)",
    "CNN\n(VGG + XGBoost)",
    "Reasoning-based VLM\n(GPT-4.1, Few-shot CoT)"
]

# Normalize each matrix by row
normalized_cms = [cm.astype('float') / cm.sum(axis=1, keepdims=True) for cm in cms]

# Set up 2x2 subplot grid without colorbars
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes = axes.flatten()

for i, ax in enumerate(axes):
    sns.heatmap(normalized_cms[i], annot=True, fmt=".2f", cmap='Blues',
                xticklabels=['Predicted: Yes', 'Predicted: No'],
                yticklabels=['Actual: Yes', 'Actual: No'],
                cbar=False, vmin=0, vmax=1, ax=ax)
    ax.set_title(titles[i], fontsize=10)
    ax.set_xlabel('')
    ax.set_ylabel('')

# Global layout
fig.suptitle("Normalized Confusion Matrices for Safety Barrier Detection", fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('RES3.png', dpi=2000)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Arial"

# Data: model -> {prompt_combo: accuracy}
data = {
    "4-0-mini": {
        "T1Z1": 0.93, "T1Z2": 0.95,
        "T2Z1": 0.84, "T2Z2": 0.79
    },
    "4-0": {
        "T1Z1": 0.95, "T1Z2": 0.95,
        "T2Z1": 0.97, "T2Z2": 0.84
    },
    "4-1-mini": {
        "T1Z1": 0.98, "T1Z2": 0.93,
        "T2Z1": 0.99, "T2Z2": 0.81
    },
    "4-1": {
        "T1Z1": 0.90, "T1Z2": 0.88,
        "T2Z1": 0.99, "T2Z2": 0.91
    }
}

# Define the x-axis order
prompt_order = ["T1Z1", "T1Z2", "T2Z1", "T2Z2"]

# Plot
plt.figure(figsize=(8, 5))

for model, results in data.items():
    accuracies = [results.get(k, None) for k in prompt_order]
    plt.plot(prompt_order, accuracies, marker='o', label=model)

plt.title("Accuracy Comparison Across VLM Models and Prompt Templates", fontsize=14, fontweight='bold')
plt.xlabel("Task-Prompt Combination", fontsize=12,fontweight='bold')
plt.ylabel("Accuracy", fontsize=12,fontweight='bold')
plt.ylim(0.75, 1.0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Model Variant")
plt.tight_layout()
plt.savefig('RES5.png', dpi=2000)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set font (optional, skip if Arial not available)
plt.rcParams["font.family"] = "Arial"

# Data
tasks = ["Task 2: Rigid Object Detection", "Task 1: Safety Barrier Detection"]
before = [0.53, 0.67]
after = [0.92, 0.82]

# Y positions
y_pos = np.arange(len(tasks)) * 0.6
bar_width = 0.25

# Plot
plt.figure(figsize=(10, 4))
bars1 = plt.barh(y_pos - bar_width/2, before, height=bar_width, label='Hand-crafted Captions', color='#d62728')
bars2 = plt.barh(y_pos + bar_width/2, after, height=bar_width, label='Optimized with CoOp', color='#2ca02c')

# Axes and labels
plt.yticks(y_pos, tasks, fontweight='bold')
plt.xlim(0.4, 1.0)
plt.xlabel("Average Accuracy", fontsize=12, fontweight='bold')
plt.title("Performance Gain in CLIP via Caption Optimization Using CoOp", fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Annotate bars
for bars in [bars1, bars2]:
    for bar in bars:
        width = bar.get_width()
        plt.text(width + 0.01, bar.get_y() + bar.get_height()/2, f"{width:.2f}", va='center')

# Legend and layout
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('RES6.png', dpi=2000)

plt.show()
```
